Метод ковзного допуску входить до методу штрафних функцій. Він дозволяє покращити значення цільової функції як за рахунок інформації, отриманої в допустимих точках, так і за рахунок інформації, що вдається отримати при проходженні через деякі точки, що лежать за межами допустимої області(майже допустимі точки). Інтервали, при яких точки можна вважати майже допустимими в ході оптимізації поступово скорочуються. При наближенні до оптимальної точки враховуються лише строго допустимі точки. 
З такою стратегією пошуку занальну задачу нелінійного програмування можна замінити більш простою: 
мінімізувати функцію

$z = f(x), x\in R^n$

При одному обмеженні:

$\Phi ^{(k)} - T(x)\geq 0$

Завдяки цим двом функція можна зрозуміти, чи функціям є допустимою:

- точка $x ^{(k)}$ допустима, якщо $T(x^{(k)}) = 0$;
- точка $x ^{(k)}$ майже допустима, якщо $0 \leq T(x^{(k)}) \leq \Phi^{(k)}$;
- точка $x ^{(k)}$ недопустими, якщо $T(x^{(k)}) > \Phi^{(k)}$.

Робота алгоритму може бути завершена в двох випадках, а саме:
- коли $\Phi^{(k)} \leq \varepsilon$. В даному випадку, алгоритм вважається успішно завершеним.
- коли не вдається знайти допустиму чи майже допустиму точку. В цьому випідку необхідно почати пошук спочатку, з нової стартової точки.

Перейдемо до методу Нелдера-Міда. 
Цей метод є методом нульового порядку(тобто, не потребує обчислення похідних для розв'язку), а таком працює на основі методу симплексу. 

Метод Нелдера-Міда використовує можливість розтягнення або стиснення симплексу при його відображенні. 

Для пошуку точки в межаї цього методу використовуються вершини симплексу:
- $x^{(h)}$ - вершина з найбільшим значенням функції;
- $x^{(g)}$ - вершина, якій відповідає наступне по величині значення функції
- $x^{(l)}$ - вершина із найменшим значенням функції.

Відображення вершини симплексу виконується вздовж прямої $x = x^{(h)} + \lambda(x_c-x^{(h)})$, де $\lambda = 1 + \theta$.

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

## Створення функцій для обчислення

У наступному коді буде створено цільову функцію задачі та функції обмежень, а також, функція $T(x)$, що була виведена аналітично, з використання раніше згаданих формул.

```python
def func(x):
    x_1, x_2 = x
    return (10 * (x_1 - x_2) ** 2 + (x_1 - 1) ** 2) ** 4


def lim_1(x):
    x_1, x_2 = x
    return x_1


def lim_2(x):
    x_1, x_2 = x
    return x_2


def lim_3(x):
    x_1, x_2 = x
    return -x_1 - x_2 + 4


def t_x(x, u):
    x_1, x_2 = x
    u_1, u_2, u_3 = u
    return np.sqrt(u_1 * x_1**2 + u_2 * x_2**2 + u_3 * (-x_1 - x_2 + 4)**2)


def lim_32(x):
    x_1, x_2 = x
    return -2 * x_1 - x_2 + 2


def lim_33(x):
    x_1, x_2 = x
    return - x_1 ** 2 - x_2 ** 2 + 4


def t_x2(x, u):
    x_1, x_2 = x
    u_1, u_2, u_3 = u
    return np.sqrt(u_1 * x_1**2 + u_2 * x_2**2 + u_3 * (-2*x_1 - x_2 + 2)**2)


def t_x3(x, u):
    x_1, x_2 = x
    u_1, u_2, u_3 = u
    return np.sqrt(u_1 * x_1**2 + u_2 * x_2**2 + u_3 * (- x_1 ** 2 - x_2 ** 2 + 4))

```

## Визначення функцій

В даній роботі буде використовуватися дві основні функції, назви яких відповідають методам, що досліджуються в курсовій роботі, а саме: *nelder_mead()* та *sliding_tolerance()*.


```python
def nelder_mead(x_0, x_1, x_2, f_count):
    # Спершу побудуємо симплекс, для чого знайдемо значення d_1, d_2
    global x_h, f_h, f_g, f_l, teta, x_l, x_g
    n = 2
    epsilon = 0.0001
    alpha = 1
    beta = 0.5
    gamma = 2
    f_0 = func(x_0)
    f_1 = func(x_1)
    f_2 = func(x_2)
    if f_0 > f_1 and f_0 > f_2:
        x_h = x_0
        f_h = f_0
        if f_1 > f_2:
            x_g = x_1
            f_g = f_1
            x_l = x_2
            f_l = f_2

        else:
            x_g = x_2
            f_g = f_2
            x_l = x_1
            f_l = f_1
    elif f_1 > f_0 and f_1 > f_2:
        x_h = x_1
        f_h = f_1
        if f_0 > f_2:
            x_g = x_0
            f_g = f_0
            x_l = x_2
            f_l = f_2
        else:
            x_g = x_2
            f_g = f_2
            x_l = x_0
            f_l = f_0
    elif f_2 > f_0 and f_2 > f_1:
        x_h = x_2
        f_h = f_2
        if f_0 > f_1:
            x_g = x_0
            f_g = f_0
            x_l = x_1
            f_l = f_1
        else:
            x_g = x_1
            f_g = f_1
            x_l = x_0
            f_l = f_0
    # print(x_h, x_g, x_l)
    # print(f_h, f_g, f_l)
    x_c = [(x_l[0] + x_g[0])/2, (x_l[1] + x_g[1])/2]
    f_c = func(x_c)
    # print(x_c)
    x_new = [2*x_c[0] - x_h[0], 2*x_c[1] - x_h[1]]
    f_new = func(x_new)
    # print(f_new)
    if f_new > f_h:
        teta = -0.5
    elif f_g < f_new < f_h:
        teta = beta
    elif f_g >= f_new >= f_l:
        teta = alpha
    elif f_new < f_l:
        teta = gamma
    koef = 1 + teta
    x_new = [x_h[0] + koef*(x_c[0] - x_h[0]), x_h[0] + koef*(x_c[0] - x_h[0])]
    f_new = func(x_new)
    func_means = [f_l, f_g, f_h]
    sum = 0
    for elm in func_means:
        sum += (elm - f_c) ** 2
    stop = sum / (n+1)
    if stop <= epsilon:
        return x_g, x_l, x_new, x_c, f_count + 9
    else:
        # print('recurtion')
        return nelder_mead(x_g, x_l, x_new, f_count + 9)
```

Також представлена функція модивікованого методу Недлера-Міда, де є допустима перевірка ще одразу після пробного відображення.

```python
def nelder_mead_modified(x_0, x_1, x_2, f_count):
    # Спершу побудуємо симплекс, для чого знайдемо значення d_1, d_2
    global x_h, f_h, f_g, f_l, teta, x_l, x_g
    n = 2
    epsilon = 0.0001
    alpha = 1
    beta = 0.5
    gamma = 2
    f_0 = func(x_0)
    f_1 = func(x_1)
    f_2 = func(x_2)
    if f_0 > f_1 and f_0 > f_2:
        x_h = x_0
        f_h = f_0
        if f_1 > f_2:
            x_g = x_1
            f_g = f_1
            x_l = x_2
            f_l = f_2
        else:
            x_g = x_2
            f_g = f_2
            x_l = x_1
            f_l = f_1
    elif f_1 > f_0 and f_1 > f_2:
        x_h = x_1
        f_h = f_1
        if f_0 > f_2:
            x_g = x_0
            f_g = f_0
            x_l = x_2
            f_l = f_2
        else:
            x_g = x_2
            f_g = f_2
            x_l = x_0
            f_l = f_0
    elif f_2 > f_0 and f_2 > f_1:
        x_h = x_2
        f_h = f_2
        if f_0 > f_1:
            x_g = x_0
            f_g = f_0
            x_l = x_1
            f_l = f_1
        else:
            x_g = x_1
            f_g = f_1
            x_l = x_0
            f_l = f_0
    # print(x_h, x_g, x_l)
    # print(f_h, f_g, f_l)
    x_c = [(x_l[0] + x_g[0])/2, (x_l[1] + x_g[1])/2]
    f_c = func(x_c)
    # print(x_c)
    x_new = [2*x_c[0] - x_h[0], 2*x_c[1] - x_h[1]]
    f_new = func(x_new)
    # print(f_new)
    if f_new > f_h:
        teta = -0.5
    elif f_g < f_new < f_h:
        teta = beta
    elif f_g >= f_new >= f_l:
        teta = alpha
    elif f_new < f_l:
        teta = gamma
    koef = 1 + teta
    x_new = [x_h[0] + koef*(x_c[0] - x_h[0]), x_h[0] + koef*(x_c[0] - x_h[0])]
    f_new = func(x_new)
    sum = 0
    func_means = [f_l, f_g, f_h]
    for elm in func_means:
        sum += (elm - f_c) ** 2
    stop = sum / (n + 1)
    if stop <= epsilon:
        return x_g, x_l, x_new, x_c, f_count + 9
    func_means = [f_l, f_g, f_h]
    sum = 0
    for elm in func_means:
        sum += (elm - f_c) ** 2
    stop = sum / (n+1)
    if stop <= epsilon:

        return x_g, x_l, x_new, x_c, f_count + 9
    else:
        print('recurtion')
        return nelder_mead_modified(x_g, x_l, x_new, f_count + 9)
```

```python
def sliding_tolerance(x_0, x_1, x_2, PHI_k, r, m, search, f_count):
    # search = []
    epsilon = 0.0001
    x_1, x_2, x_3, x_c, f_count = nelder_mead(x_0, x_1, x_2, f_count)
    x_list = [x_1, x_2, x_3]
    sum = 0
    for elm in x_list:
        sum_elm = [elm[0] - x_c[0], elm[1] - x_c[1]]
        sum += LA.norm(sum_elm)
    phi_k = (m + 1) * sum / (r + 1)
    PHI_k.append(phi_k)
    u = []
    g_1 = lim_1(x_3)
    g_2 = lim_2(x_3)
    g_3 = lim_3(x_3)
    g_list = [g_1, g_2, g_3]
    for elm in g_list:
        if elm < 0:
            u.append(1)
        else:
            u.append(0)

    T_x = t_x(x_3, u)
    phi_k = min(PHI_k)
    if T_x <= phi_k <= epsilon:
        search.append(x_3)
        return x_3, func(x_3), search, f_count
    else:
        search.append(x_3)
        return sliding_tolerance(x_1, x_2, x_3, PHI_k, r, m, search, f_count)
```

## Обчислення

У наступному шматкові коду визначається точка $x^{(0)}$, обчислюється початкой симплекс, також створюються смінні розмірності, відстані між вершинами симплексу, кількість обмежень-рівностей, кількість ступенів свободи, список зі значеннями $\Phi^{(k)}$ та список для точок пошуку.

```python
x_0 = [-1.2, 0]  # Точка, з якої починається пошук
n = 2  # Розмірність
t = 1  # відстань між вершинами
# Створення симплексу
d_1 = (t * (np.sqrt(n + 1) + n - 1)) / (n * np.sqrt(2))
d_2 = (t * (np.sqrt(n + 1) - 1)) / (n * np.sqrt(2))
x_1 = [d_1, d_2]
x_2 = [d_2, d_1]
m = 0  # кільсть обмежень, виражених рівністю
r = n - m  # кількість ступенів свободи
f_count = 0

phi_0 = 2*(m + 1)*t
PHI_k = [phi_0]
search = []

x_min, f_min, search, f_count = sliding_tolerance(x_0, x_1, x_2, PHI_k, r, m, search, f_count)
```
